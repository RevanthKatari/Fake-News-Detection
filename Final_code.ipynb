```python
import pandas as pd
import torch
import numpy as np
import cv2
import librosa
from PIL import Image
from tqdm import tqdm

from sentence_transformers import SentenceTransformer
from transformers import (
    CLIPProcessor,
    CLIPModel,
    Wav2Vec2Processor,
    Wav2Vec2Model,
    VideoMAEImageProcessor,
    VideoMAEModel
)

import timm
import torchvision.transforms as transforms
```

```python
df = pd.read_csv("balanced_multimodal_welfake.csv")

print("Balanced dataset size:", len(df))
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using:", device)
```

```python
print("Loading models...")

# TEXT
text_model = SentenceTransformer("all-mpnet-base-v2").to(device)

# IMAGE
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

dino_model = timm.create_model("vit_base_patch14_dinov2", pretrained=True)
dino_model.to(device).eval()

dino_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
])

# AUDIO
audio_processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
audio_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base").to(device)

# VIDEO
video_processor = VideoMAEImageProcessor.from_pretrained("MCG-NJU/videomae-base")
video_model = VideoMAEModel.from_pretrained("MCG-NJU/videomae-base").to(device)

print("✅ Models Loaded")
```

```python
text_features = []
image_features = []
audio_features = []
video_features = []
```

```python
def extract_text(text):
    if not isinstance(text, str) or text.strip()=="":
        return np.zeros(768)
    return text_model.encode(text)
```

```python
def extract_image(path):

    if not isinstance(path,str) or path.strip()=="":
        return np.zeros(512+768)

    try:
        image = Image.open(path).convert("RGB")

        clip_inputs = clip_processor(images=image,
                                     return_tensors="pt").to(device)

        clip_feat = clip_model.get_image_features(**clip_inputs)
        clip_feat = clip_feat.detach().cpu().numpy().flatten()

        img = dino_transform(image).unsqueeze(0).to(device)

        with torch.no_grad():
            dino_feat = dino_model.forward_features(img)

        dino_feat = dino_feat.mean(dim=1).cpu().numpy().flatten()

        return np.concatenate([clip_feat,dino_feat])

    except:
        return np.zeros(512+768)
```

```python
def extract_audio(path):

    if not isinstance(path,str) or path.strip()=="":
        return np.zeros(768)

    try:
        speech,_ = librosa.load(path,sr=16000)

        inputs = audio_processor(
            speech,
            sampling_rate=16000,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            outputs = audio_model(**inputs)

        emb = outputs.last_hidden_state.mean(dim=1)
        return emb.cpu().numpy().flatten()

    except:
        return np.zeros(768)
```

```python
def extract_video(path,num_frames=16):

    if not isinstance(path,str) or path.strip()=="":
        return np.zeros(768)

    try:
        cap=cv2.VideoCapture(path)
        frames=[]

        total=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        step=max(total//num_frames,1)

        for i in range(0,total,step):
            cap.set(cv2.CAP_PROP_POS_FRAMES,i)
            ret,frame=cap.read()
            if not ret:
                break

            frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            frames.append(frame)

            if len(frames)==num_frames:
                break

        cap.release()

        if len(frames)==0:
            return np.zeros(768)

        inputs=video_processor(frames,
                               return_tensors="pt").to(device)

        with torch.no_grad():
            outputs=video_model(**inputs)

        emb=outputs.last_hidden_state.mean(dim=1)

        return emb.cpu().numpy().flatten()

    except:
        return np.zeros(768)
```

```python
for _,row in tqdm(df.iterrows(),total=len(df)):

    text_features.append(
        extract_text(row.get("text",""))
    )

    image_features.append(
        extract_image(row.get("image_path",""))
    )

    audio_features.append(
        extract_audio(row.get("audio_path",""))
    )

    video_features.append(
        extract_video(row.get("video_path",""))
    )
```

```python
np.save("text_embeddings.npy",np.array(text_features))
np.save("image_embeddings.npy",np.array(image_features))
np.save("audio_embeddings.npy",np.array(audio_features))
np.save("video_embeddings.npy",np.array(video_features))

# Save labels + metadata alignment
df.to_csv("embedding_metadata.csv",index=False)

print("✅ Embeddings Saved Successfully")
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# ======================================================
# DEVICE
# ======================================================

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# ======================================================
# LOAD EMBEDDINGS
# ======================================================

text = np.load("text_embeddings.npy")
image = np.load("image_embeddings.npy")
audio = np.load("audio_embeddings.npy")
video = np.load("video_embeddings.npy")
labels = np.load("labels.npy")

print("Loaded embeddings:")
print(text.shape, image.shape, audio.shape, video.shape)

# ======================================================
# DATASET
# ======================================================

class MultimodalDataset(Dataset):
    def __init__(self, t, i, a, v, y):
        self.text = torch.tensor(t, dtype=torch.float32)
        self.image = torch.tensor(i, dtype=torch.float32)
        self.audio = torch.tensor(a, dtype=torch.float32)
        self.video = torch.tensor(v, dtype=torch.float32)
        self.labels = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return (
            self.text[idx],
            self.image[idx],
            self.audio[idx],
            self.video[idx],
            self.labels[idx],
        )

# ======================================================
# TRAIN TEST SPLIT
# ======================================================

X_train, X_test, I_train, I_test, A_train, A_test, V_train, V_test, y_train, y_test = train_test_split(
    text, image, audio, video, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

train_dataset = MultimodalDataset(X_train, I_train, A_train, V_train, y_train)
test_dataset = MultimodalDataset(X_test, I_test, A_test, V_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)

# ======================================================
# CROSS ATTENTION MODEL
# ======================================================
class CrossAttentionModel(nn.Module):

    def __init__(self, text_dim, image_dim, audio_dim, video_dim, hidden=512):
        super().__init__()

        # Projection layers
        self.text_proj = nn.Linear(text_dim, hidden)
        self.image_proj = nn.Linear(image_dim, hidden)
        self.audio_proj = nn.Linear(audio_dim, hidden)
        self.video_proj = nn.Linear(video_dim, hidden)

        # Cross Attention
        self.cross_attn = nn.MultiheadAttention(
            embed_dim=hidden,
            num_heads=8,
            batch_first=True
        )

        # Classifier
        self.classifier = nn.Sequential(
            nn.Linear(hidden, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 2)
        )

    def forward(self, text, image, audio, video):

        # Project to same dimension
        t = self.text_proj(text).unsqueeze(1)
        i = self.image_proj(image).unsqueeze(1)
        a = self.audio_proj(audio).unsqueeze(1)
        v = self.video_proj(video).unsqueeze(1)

        # Concatenate non-text modalities
        other_modalities = torch.cat([i, a, v], dim=1)

        # STRICT CROSS ATTENTION
        fused, _ = self.cross_attn(
            query=t,
            key=other_modalities,
            value=other_modalities
        )

        fused = fused.squeeze(1)

        out = self.classifier(fused)

        return out

# ======================================================
# MODEL INIT
# ======================================================

model = CrossAttentionModel(
    text.shape[1],
    image.shape[1],
    audio.shape[1],
    video.shape[1]
).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# ======================================================
# TRAINING LOOP
# ======================================================

EPOCHS = 10

for epoch in range(EPOCHS):

    model.train()
    total_loss = 0

    for t, i, a, v, y in train_loader:

        t, i, a, v, y = (
            t.to(device),
            i.to(device),
            a.to(device),
            v.to(device),
            y.to(device),
        )

        optimizer.zero_grad()

        outputs = model(t, i, a, v)
        loss = criterion(outputs, y)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1} Loss: {total_loss:.4f}")

# ======================================================
# EVALUATION
# ======================================================

model.eval()

preds = []
truths = []

with torch.no_grad():
    for t, i, a, v, y in test_loader:

        t, i, a, v = (
            t.to(device),
            i.to(device),
            a.to(device),
            v.to(device),
        )

        outputs = model(t, i, a, v)
        pred = torch.argmax(outputs, dim=1).cpu().numpy()

        preds.extend(pred)
        truths.extend(y.numpy())

print("\nAccuracy:", accuracy_score(truths, preds))
print("\nClassification Report:")
print(classification_report(truths, preds))

# ======================================================
# SAVE MODEL
# ======================================================

torch.save(model.state_dict(), "cross_attention_model.pth")

print("\n✅ Model saved!")
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# ======================================================
# DEVICE
# ======================================================

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# ======================================================
# LOAD EMBEDDINGS
# ======================================================

text = np.load("text_embeddings.npy")
image = np.load("image_embeddings.npy")
audio = np.load("audio_embeddings.npy")
video = np.load("video_embeddings.npy")
labels = np.load("labels.npy")

print("Loaded embeddings:")
print(text.shape, image.shape, audio.shape, video.shape)

# ======================================================
# DATASET
# ======================================================

class MultimodalDataset(Dataset):

    def __init__(self, t, i, a, v, y):

        self.text = torch.tensor(t, dtype=torch.float32)
        self.image = torch.tensor(i, dtype=torch.float32)
        self.audio = torch.tensor(a, dtype=torch.float32)
        self.video = torch.tensor(v, dtype=torch.float32)
        self.labels = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):

        return (
            self.text[idx],
            self.image[idx],
            self.audio[idx],
            self.video[idx],
            self.labels[idx]
        )

# ======================================================
# TRAIN TEST SPLIT
# ======================================================

X_train, X_test, I_train, I_test, A_train, A_test, V_train, V_test, y_train, y_test = train_test_split(
    text,
    image,
    audio,
    video,
    labels,
    test_size=0.2,
    stratify=labels,
    random_state=42
)

train_dataset = MultimodalDataset(X_train, I_train, A_train, V_train, y_train)
test_dataset = MultimodalDataset(X_test, I_test, A_test, V_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)

# ======================================================
# GATED FUSION MODEL
# ======================================================

class GatedFusionModel(nn.Module):

    def __init__(self, text_dim, image_dim, audio_dim, video_dim, hidden=512):

        super().__init__()

        # Projection layers
        self.text_proj = nn.Linear(text_dim, hidden)
        self.image_proj = nn.Linear(image_dim, hidden)
        self.audio_proj = nn.Linear(audio_dim, hidden)
        self.video_proj = nn.Linear(video_dim, hidden)

        # Gate layers
        self.text_gate = nn.Linear(hidden, hidden)
        self.image_gate = nn.Linear(hidden, hidden)
        self.audio_gate = nn.Linear(hidden, hidden)
        self.video_gate = nn.Linear(hidden, hidden)

        self.sigmoid = nn.Sigmoid()

        # Classifier
        self.classifier = nn.Sequential(
            nn.Linear(hidden, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 2)
        )

    def forward(self, text, image, audio, video):

        # Projection
        t = self.text_proj(text)
        i = self.image_proj(image)
        a = self.audio_proj(audio)
        v = self.video_proj(video)

        # Gates
        gt = self.sigmoid(self.text_gate(t))
        gi = self.sigmoid(self.image_gate(i))
        ga = self.sigmoid(self.audio_gate(a))
        gv = self.sigmoid(self.video_gate(v))

        # Apply gates
        t = gt * t
        i = gi * i
        a = ga * a
        v = gv * v

        # Fusion
        fused = (t + i + a + v) / 4

        # Classification
        out = self.classifier(fused)

        return out

# ======================================================
# MODEL INIT
# ======================================================

model = GatedFusionModel(
    text.shape[1],
    image.shape[1],
    audio.shape[1],
    video.shape[1]
).to(device)

# Class weights for imbalance
class_weights = torch.tensor([1.7, 0.7]).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# ======================================================
# TRAINING
# ======================================================

EPOCHS = 10

for epoch in range(EPOCHS):

    model.train()
    total_loss = 0

    for t, i, a, v, y in train_loader:

        t = t.to(device)
        i = i.to(device)
        a = a.to(device)
        v = v.to(device)
        y = y.to(device)

        optimizer.zero_grad()

        outputs = model(t, i, a, v)

        loss = criterion(outputs, y)

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1} Loss: {total_loss:.4f}")

# ======================================================
# EVALUATION
# ======================================================

model.eval()

preds = []
truths = []

with torch.no_grad():

    for t, i, a, v, y in test_loader:

        t = t.to(device)
        i = i.to(device)
        a = a.to(device)
        v = v.to(device)

        outputs = model(t, i, a, v)

        pred = torch.argmax(outputs, dim=1).cpu().numpy()

        preds.extend(pred)
        truths.extend(y.numpy())

print("\nAccuracy:", accuracy_score(truths, preds))

print("\nClassification Report:")
print(classification_report(truths, preds))

# ======================================================
# SAVE MODEL
# ======================================================

torch.save(model.state_dict(), "gated_fusion_model.pth")

print("\nModel saved successfully!")
```

```python
# =========================================
# 1. IMPORTS
# =========================================
import pandas as pd
import torch
import torch.nn as nn
import shap
from sentence_transformers import SentenceTransformer
import random

# =========================================
# 2. DEVICE SETUP
# =========================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# =========================================
# 3. LOAD & CLEAN DATASET (FIXED)
# =========================================
df = pd.read_csv("balanced_multimodal_welfake.csv")

# Remove actual NaN values
df = df.dropna(subset=['text'])

# Convert to string
df['text'] = df['text'].astype(str)

# Remove invalid text entries
df = df[
    (df['text'].str.strip() != "") &
    (df['text'].str.lower() != "nan")
]

# Remove very short text (important for SHAP)
df = df[df['text'].str.len() > 20]

print("Clean dataset size:", len(df))

# =========================================
# 4. LOAD SBERT MODEL
# =========================================
sbert_model = SentenceTransformer('all-MiniLM-L6-v2', device=device)

# =========================================
# 5. DEFINE CLASSIFIER
# =========================================
class Classifier(nn.Module):
    def __init__(self, input_dim=384, hidden_dim=128, num_classes=2):
        super(Classifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x

classifier = Classifier().to(device)

# 🔴 LOAD YOUR TRAINED MODEL HERE
# classifier.load_state_dict(torch.load("model.pth", map_location=device))

classifier.eval()

# =========================================
# 6. PREDICTION FUNCTION (SAFE)
# =========================================
def predict(texts):
    # Extra safety cleaning
    texts = [
        str(t)[:300] if t and str(t).strip().lower() != "nan" else ""
        for t in texts
    ]

    embeddings = sbert_model.encode(
        texts,
        convert_to_tensor=True,
        device=device
    )

    with torch.no_grad():
        outputs = classifier(embeddings)
        probs = torch.softmax(outputs, dim=1)

    return probs.cpu().numpy()

# =========================================
# 7. SHAP EXPLAINER (STABLE SETTINGS)
# =========================================
explainer = shap.Explainer(
    predict,
    masker=shap.maskers.Text(),
    max_evals=20
)

# =========================================
# 8. SAFE SAMPLING
# =========================================
fake_texts = df[df['label'] == 1]['text'].tolist()
real_texts = df[df['label'] == 0]['text'].tolist()

def get_valid_samples(text_list, n=2):
    valid = [t for t in text_list if isinstance(t, str) and len(t.strip()) > 20]
    return random.sample(valid, min(n, len(valid)))

fake_samples = get_valid_samples(fake_texts)
real_samples = get_valid_samples(real_texts)

# =========================================
# 9. EXPLAIN FUNCTION (SAFE)
# =========================================
def explain_text(text, title=""):
    # Final safety check
    if text is None or str(text).strip() == "" or str(text).lower() == "nan":
        print("Skipping invalid text")
        return

    print("\n" + "="*60)
    print(title)
    print("="*60)
    print("\nTEXT:\n", text[:500])

    try:
        shap_values = explainer([text])
        shap.plots.text(shap_values[0])
    except Exception as e:
        print("SHAP failed for this sample:", e)

# =========================================
# 10. RUN EXPLANATIONS
# =========================================
print("\n========== FAKE NEWS EXAMPLES ==========")
for i, txt in enumerate(fake_samples):
    explain_text(txt, f"Fake Sample {i+1}")

print("\n========== REAL NEWS EXAMPLES ==========")
for i, txt in enumerate(real_samples):
    explain_text(txt, f"Real Sample {i+1}")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn

# =========================================
# LOAD DATA
# =========================================
df = pd.read_csv("balanced_multimodal_welfake.csv")

# If label is string, convert to numeric
if df["label"].dtype == object:
    df["label_num"] = df["label"].str.lower().map({"real": 0, "fake": 1})
else:
    df["label_num"] = df["label"].astype(int)

# =========================================
# LOAD EMBEDDINGS + MODEL
# =========================================
text_emb = np.load("text_embeddings.npy")
image_emb = np.load("image_embeddings.npy")
audio_emb = np.load("audio_embeddings.npy")
video_emb = np.load("video_embeddings.npy")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class MultimodalClassifier(nn.Module):
    def __init__(self, text_dim=768, image_dim=1280, audio_dim=768, video_dim=768, hidden_dim=256):
        super().__init__()
        self.text_proj = nn.Linear(text_dim, 128)
        self.image_proj = nn.Linear(image_dim, 128)
        self.audio_proj = nn.Linear(audio_dim, 128)
        self.video_proj = nn.Linear(video_dim, 128)

        self.classifier = nn.Sequential(
            nn.Linear(128 * 4, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_dim, 2)
        )

    def forward(self, t, i, a, v):
        t = self.text_proj(t)
        i = self.image_proj(i)
        a = self.audio_proj(a)
        v = self.video_proj(v)

        x = torch.cat([t, i, a, v], dim=1)
        return self.classifier(x)

model = MultimodalClassifier().to(device)
model.load_state_dict(torch.load("multimodal_model.pth", map_location=device))
model.eval()

# =========================================
# PREDICTION FUNCTION
# =========================================
def predict_sample(idx):
    t = torch.tensor(text_emb[idx], dtype=torch.float32).unsqueeze(0).to(device)
    i = torch.tensor(image_emb[idx], dtype=torch.float32).unsqueeze(0).to(device)
    a = torch.tensor(audio_emb[idx], dtype=torch.float32).unsqueeze(0).to(device)
    v = torch.tensor(video_emb[idx], dtype=torch.float32).unsqueeze(0).to(device)

    with torch.no_grad():
        logits = model(t, i, a, v)
        probs = torch.softmax(logits, dim=1).cpu().numpy()[0]
        pred = int(np.argmax(probs))

    return pred, probs

# =========================================
# SHOW ONE FAKE + ONE REAL PER MODALITY GROUP
# =========================================
def show_examples_for_group(group_name):
    print("\n" + "="*80)
    print(f"MODALITY GROUP: {group_name}")
    print("="*80)

    subset = df[df["modality_group"] == group_name]

    for true_label, label_name in [(1, "FAKE"), (0, "REAL")]:
        row = subset[subset["label_num"] == true_label].sample(1, random_state=42)
        idx = row.index[0]

        pred, probs = predict_sample(idx)

        print("\n" + "-"*60)
        print(f"Ground Truth: {label_name}")
        print(f"Predicted   : {'FAKE' if pred == 1 else 'REAL'}")
        print(f"Confidence  : REAL={probs[0]:.4f}, FAKE={probs[1]:.4f}")
        print(f"Correct?    : {'YES' if pred == true_label else 'NO'}")
        print(f"Dataset idx : {idx}")

        text_value = str(df.loc[idx, "text"])[:700]
        print("\nTEXT:")
        print(text_value if text_value.strip() else "[No text available]")

        print("\nIMAGE PATH:")
        print(df.loc[idx, "image_path"])

        print("\nAUDIO PATH:")
        print(df.loc[idx, "audio_path"])

        print("\nVIDEO PATH:")
        print(df.loc[idx, "video_path"])

# Run for each group
show_examples_for_group("text_only")
show_examples_for_group("text_image")
show_examples_for_group("audio_video")
```
